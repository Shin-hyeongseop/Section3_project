```python
df = pd.read_csv('p3data.csv', sep=',')
df
```

```python
df['Volume'] = df['Volume'] * (1/100)
```

```python
df['Date'] = df['Date'].str.replace('-','',)
```

```python
df = sklearn.utils.shuffle(df)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, random_state=2)
```

```python
train
```

```python
target = 'Close'
features = df.columns.drop([target])
X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]
```

```python
train['Close'].mean()
```

```python
predict = y_train.mean()
```

```python
predict
```

```python
#   
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
y_pred = [predict] * len(y_train)
mae = mean_absolute_error(y_train, y_pred)
mse = mean_squared_error(y_train, y_pred)
rmse = mse * 0.5
r2 = r2_score(y_train, y_pred)


mae, mse, rmse, r2
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state = 2, n_estimators = 3,  min_samples_leaf = 4, min_samples_split = 20, max_depth = 2)
rf_model.fit(X_train, y_train)
```

```python
models = ['rf']
mse, rmse, mae, r2 = [], [], [], []
```

```python
rf_y_pred = rf_model.predict(X_train)
mae = mean_absolute_error(y_train, rf_y_pred)
mse = mean_squared_error(y_train, rf_y_pred)
rmse = mse * 0.5
r2 = r2_score(y_train, rf_y_pred)

mae, mse, rmse, r2
```

```python
X_total = pd.concat([X_train, X_test])
y_total = pd.concat([y_train, y_test])
```

```python
rf_model = RandomForestRegressor(random_state = 42, n_estimators = 10,  min_samples_leaf = 8, min_samples_split = 20, max_depth = 4)
rf_model.fit(X_total, y_total)
```

```python
rf_y_pred = rf_model.predict(X_total)
mae = mean_absolute_error(y_total, rf_y_pred)
mse = mean_squared_error(y_total, rf_y_pred)
rmse = mse * 0.5
r2 = r2_score(y_total, rf_y_pred)

mae, mse, rmse, r2
```
